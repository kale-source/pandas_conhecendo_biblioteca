Qual o valor médio de aluguel por tipo de imóvel?



```python
dados.head()
```

```python
dados['Valor'].mean()
```

```python
# Funcionna muito parecido com o group by do sql
# Faz o agrupamento de cada tipo e transforma em um dataframe
# Através disso podemos saber a média de cada coluna agrupado por cada tipo de imovel
dados.groupby('Tipo').mean(numeric_only=True)
```

```python
# Agora queremos saber qual é a média da coluna Valor
dados.groupby('Tipo')['Valor'].mean() 
# Porém não trouxe em um dataframe, então temos que colocar em dataframe
```

```python
# Definimos o dataframe colocando o ['Valor'] dentro de outros colchetes [['Valor']] e ordenamos em ordem crescente com sort_values() especificando a coluna 'Valor'
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# Gerando gráfico
values = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
values.plot(kind='barh', figsize=(14, 10), color='green')
```

REMOVENDO IMÓVEIS COMERCIAIS

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']

dados.query('@imoveis_comerciais not in Tipo')
```

```python
df = dados.query('@imoveis_comerciais not in Tipo')
df.Tipo.unique()
```

```python
df_group = df.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor')

df_group.plot(kind='barh', figsize=(14, 10), color='orange')
```

Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df.Tipo.value_counts(normalize=True)
```

```python
df.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
```

```python
df_percentage = df.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
df_percentage.rename(columns={'proportion': 'Percentual'}, inplace=True) # Mudando nome da coluna
df_percentage.plot(kind='barh', figsize=(14, 10), color='green', xlabel='Percentage', ylabel='Tipo')
```

Apesar de ser possivel substituir os valores nulos usando fillna(), podemos remover as linhas usando dropna(), basicamente tem a função de remover linhas que possuem qualquer row/column nula.

Para o método fillna, é possivel usar method=”ffill” ou method=”bfill”

serve para substituir o valor nulo com o valor anterior ou posterior.

É possível utilizar o método interpolate() para preencher os valores nulos com valores interpolados, ou seja, valores calculados a partir dos valores vizinhos.

Apartamentos que possuem 1 quarto e aluguel menor que 1200

```python
df_ap['Quartos'] == 1
# Vai retornar todas as linhas e todos aqueles que vier como True são as linhas que possuem na coluna quartos o valor 1
```

```python
selecao1 = df_ap['Quartos'] == 1
df_ap[selecao1] # Sendo assim trazendo o dataframe com os apartamentos que possuem apenas 1 quarto
```

Apartamentos que possuem pelo menos 2 quartos, aluguel menor que 3000 e area maior que 70

```python
selecao1 = df_ap['Quartos'] >= 2
selecao2 = df_ap['Valor'] < 3000
selecao3 = df_ap['Area'] > 70

selecao_final_extendida = selecao1 & selecao2 & selecao3

selecao_final_resumida = (df_ap['Quartos'] >= 2) & (df_ap['Valor'] < 3000) & (df_ap['Area'] > 70)
df_ap[selecao_final_resumida]
```
