```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url)
```

```python
alunos_df = pd.read_csv(url, sep=',')
alunos_df.isnull().sum() # Verificar se possui valores nulos
```

```python
alunos_df.fillna(0, inplace=True)
alunos_df
```

```python
alunos_df.drop([7,8], axis=0, inplace=True)
alunos_df
```

```python
aprovados = alunos_df['Aprovado'] == True
alunos_df[aprovados].replace({'Notas': 7}, 8)
```

```python
alunos_aprovados = alunos_df[aprovados].replace({'Notas': 7}, 8)
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
```
